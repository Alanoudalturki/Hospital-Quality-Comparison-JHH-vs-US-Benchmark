```python
#Create a new df for JH Hospital Medical Center
BD=pd.DataFrame(data_A, columns =['Provider ID', 'Hospital Name', 'State', 'Score', 'Mortality Rate','Measure Start Date',
                                                       'Measure End Date'] )
#Calcuate Midway Point: ( [Measure Start Date] + [Measure End Data] ) / 2  

BD['MidP'] = BD['Measure Start Date'] + (BD['Measure End Date'] - BD['Measure Start Date'])/2
BD.head()
```

```python
#Calculate average rate of MOR for JH Hospital: Rate dervies from score that is in percentages already in provided data
BD['AvgRate_JH']= data_A['Mortality Rate'].mean()

#Calcuate Rate of Vaccination for NY Presebrterian Hosp : Rate dervies from score that is in percentages already in provided data
BD['AvgRate_COM']= data_B['Mortality Rate'].mean()

#Calcuate Standards Dev for Rate of Vaccination for NY Presebryterian Hosp
BD['STDDEV_COM']= data_B['Mortality Rate'].std()

#Z-score
BD['ztsat'] = 1.96
BD.head()

#Calculate the UCL and LCL based on NY Presbyterian Hosp
BD['UCL'] = BD['AvgRate_COM'] + BD['ztsat'] * BD['STDDEV_COM']
BD['LCL'] = BD['AvgRate_COM'] - BD['ztsat'] * BD['STDDEV_COM']
BD
```

```python
#Plot Data
plt.figure(figsize=(15,5))
plt.plot('MidP', 'Mortality Rate', data=BD, marker = 's', markerfacecolor = 'blue', linewidth =1, color = 'blue')
plt.plot('MidP', 'UCL', data=BD, markersize = 0, color = 'red', linewidth =1)
plt.plot('MidP', 'LCL', data=BD, markersize = 0, color = 'red', linewidth =1)
plt.rotation=60
plt.legend(loc="lower right")
plt.title('The Jonhs Hopkins Hospital Vs. US Best Hospitals Competitors- HF Rate of Mortality Rate ')
plt.xlabel('Time')
plt.ylabel('Mortality Rate')
```
